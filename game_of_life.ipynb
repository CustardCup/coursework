```python
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
# Create a randomly generated starting board
columns = 8**2
rows = 8**2
board = np.floor(2*np.random.rand(columns+1, rows+1))
```

```python
# Update the board through iterations
x = 0
while x < 100:
  next = np.zeros((columns+1,rows+1))
  for i in range(1,rows):
    for j in range(1,columns):
      neighbourSum = 0

      if board[i-1,j-1] == 1:
        neighbourSum+=1
      if board[i-1,j  ] == 1:
        neighbourSum+=1
      if board[i-1,j+1] == 1:
        neighbourSum+=1

      if board[i,j-1] == 1:
        neighbourSum+=1
      if board[i,j+1] == 1:
        neighbourSum+=1

      if board[i+1,j-1] == 1:
        neighbourSum+=1
      if board[i+1,j  ] == 1:
        neighbourSum+=1
      if board[i+1,j+1] == 1:
        neighbourSum+=1

      if neighbourSum >= 4:       # overpopulation
        next[i,j] = 0
      else:
        if board[i,j]:           # if alive
          if neighbourSum >= 2:
            next[i,j] = 1
          else:
            next[i,j] = 0
        else:                     # if dead
          if neighbourSum == 3:
            next[i,j] = 1
          else:
            next[i,j] = 0

  # I have no animation skillz so this is the best I can do atm
  plt.imshow(next)
  plt.show()

  board = next
  x+=1
  time.sleep(1)
```
